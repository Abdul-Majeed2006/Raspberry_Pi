When you unplug the USB, your code usually stops. This notebook explains how to make your robot run automatically on battery power.

## 1. The Startup Sequence
Every time the Raspberry Pi Pico powers up, it automatically performs two searches on its internal flash drive:
1.  **`boot.py`**: It runs this first (mostly for advanced setup).
2.  **`main.py`**: It runs this second. **This is where your project logic lives.**

## 2. How to "Deploy" your Robot ðŸ¤–
If you have a script finished (e.g., `truck_code.py`) and you want to make it permanent:
1.  Save your finalized code to the Pico's storage using the MicroPico extension.
2.  **RENAME THE FILE TO `main.py`** on the Pico.
3.  Unplug from the computer and plug in your battery pack.

> **CRITICAL:** Since `main.py` runs on startup, always make sure you have a `try...except KeyboardInterrupt` block so you can still stop it if you plug it back into a computer later!

```python
import time

try:
    while True:
        print("Running on battery! ðŸ”‹")
        time.sleep(1)
except KeyboardInterrupt:
    print("Stopped by user")
```
