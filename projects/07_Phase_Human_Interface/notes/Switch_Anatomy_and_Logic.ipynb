## 1. Anatomy of a Switch

A mechanical switch is just a **Bridge**. 
- When it's UP: The bridge is open. No electricity flows.
- When it's DOWN: The bridge is closed. Electricity flows from one pin to the other.

### ðŸ’¡ Key Insight: No Polarity
Unlike an LED, a switch has no "Plus" or "Minus." You can plug it in either way! It just connects two points when pressed.

## 2. The "Floating Pin" Problem

Imagine a Pico pin is like a person standing in a dark room. 
- If you connect it to **3.3V**, the person sees "LIGHT" (State 1).
- If you connect it to **GND**, the person sees "DARK" (State 0).

**But what if nothing is connected?** 
Static electricity in the air, the wires, and even your hand can make the pin fluctuate randomly between 0 and 1. This is called a **Floating Pin**. 

To solve this, we use a **Pull Resistor**.

## 3. Pull-Down vs. Pull-Up (The Tug-of-War)

There are two ways to stop a pin from floating. Both involve tiny invisible resistors inside the Pico.

### â¬‡ï¸ Option A: Pull-Down (Active-High)
- **Wiring**: One side of switch to **3.3V**, other side to **GPIO**.
- **Mode**: `Pin.PULL_DOWN`.
- **Logic**: Default is **0**. When you press, it goes to **1**.

### â¬†ï¸ Option B: Pull-Up (Active-Low) â€” *THE PRO WAY*
- **Wiring**: One side of switch to **GND**, other side to **GPIO**.
- **Mode**: `Pin.PULL_UP`.
- **Logic**: Default is **1**. When you press, it goes to **0**.
> **Why use this?** It's safer. If a wire breaks and touches the metal frame of your robot (usually GND), nothing pops. It's also more resistant to electronic noise.

## 4. Edge Detection: The Moment of Impact

In engineering, we don't just care if a button is down. We care about the **change**.

1. **Rising Edge**: The exact millisecond the signal goes from 0 to 1 (Getting pressed).
2. **Falling Edge**: The exact millisecond it goes from 1 to 0 (Getting released).

**Lesson 2** in your folder shows you how to track the `last_state` to detect these edges. This is how you make a button "Toggle" a light instead of just flashing it.

## 5. Debouncing: Fixing Physics

When metal plates hit each other, they **vibrate** for a few milliseconds. To the Pico, this looks like someone pressing the button 50 times in a row fast like a machine gun.

**The Fix**: Ignore all inputs for ~200ms after the first press is confirmed. This allows the physical plates to settle down.

## 6. Interrupts (IRQ): The "Don't Miss a Thing" Mode

Usually, we check buttons in a loop: `if button.value() == 1:`. 
But what if your Pico is busy drawing a complex eye on the screen? It might miss your button press because it wasn't "looking" at that exact moment.

**Interrupts** are advanced signals that stop the CPU immediately, run a small function (the Handler), and then go back to what it was doing.
