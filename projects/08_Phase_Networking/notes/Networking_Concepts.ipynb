Why is the **Pico W** more expensive than the standard Pico? Because it has a **CYW43439** wireless chip soldered onto it! 

This chip allows your robot to:
1. Connect to your Home Wi-Fi.
2. Fetch data from the Internet (Time, Weather, Stock prices).
3. Host a web server (Control your truck from your phone!).

## 1. The Wi-Fi Handshake

Connecting to Wi-Fi is like walking into a party. 
- **SSID**: The name of the house (Your Router Name).
- **Password**: The secret phrase to get through the door.

In MicroPython, we use the `network` library to manage this.

## 2. NTP: Getting the "Real Time"

The Pico doesn't have a battery-backed clock. When you turn it off, it "forgets" what time it is. 

**NTP (Network Time Protocol)** allows the Pico to ping a server on the internet and ask: "What time is it right now?" 

The server responds with a timestamp (Seconds since Jan 1, 1970). The Pico then sets its internal clock to match.

## 3. The Architecture of a Web Request

When you want to "Search for something" or fetch data, you use an **HTTP Request**.
1. **GET**: "Please give me the data at this URL."
2. **RESPONSE**: The server sends back JSON or HTML.

We use the `urequests` library for this. It's like a tiny version of the Python `requests` library you might use on a PC.
