## 1. Mastering the Grid

Your screen is a grid of **128 (X) by 64 (Y)** pixels.
- **X (Horizontal):** 0 is the left, 127 is the right.
- **Y (Vertical):** 0 is the top, 63 is the bottom.

### Interactive Challenge: The Corners
Run this code to light up only the 4 extreme corners of your screen.

```python
from machine import Pin, I2C
from ssd1306 import SSD1306_I2C

i2c = I2C(0, sda=Pin(16), scl=Pin(17))
oled = SSD1306_I2C(128, 64, i2c)

oled.fill(0)
oled.pixel(0, 0, 1)      # Top Left
oled.pixel(127, 0, 1)    # Top Right
oled.pixel(0, 63, 1)     # Bottom Left
oled.pixel(127, 63, 1)   # Bottom Right
oled.show()
```

## 2. Drawing Shapes

The `ssd1306` driver gives us powerful tools to draw shapes without calculating every pixel.

### **The Line Tool**
`oled.line(x1, y1, x2, y2, color)`

### **The Rectangle Tools**
- `oled.rect(x, y, w, h, color)` -> Draws an **outline**.
- `oled.fill_rect(x, y, w, h, color)` -> Draws a **solid box**.

### Example: Building a Dashboard UI

```python
oled.fill(0)

# 1. Header Bar
oled.fill_rect(0, 0, 128, 12, 1)  # White bar at top
oled.text("TRUCK STATUS", 15, 2, 0) # Black text on white bar

# 2. Status Box
oled.rect(10, 20, 108, 30, 1)
oled.text("STOPPED", 35, 30, 1)

oled.show()
```

## 3. Animation: The "Movie" Concept

Animation is just drawing something, moving it slightly, and showing it again very fast. 

**The Loop Logic:**
1. `oled.fill(0)` -> Wipe the old frame.
2. Update the `x` or `y` coordinate.
3. `oled.rect(x, y, ...)` -> Draw at new position.
4. `oled.show()` -> Show the new frame.

### Run this to see a bouncing square:

```python
import time

x = 0
velocity = 2

for i in range(100): # Run for 100 frames
    oled.fill(0)
    
    # Update Position
    x += velocity
    
    # Bounce off walls
    if x > 110 or x < 0:
        velocity = -velocity
        
    # Draw
    oled.fill_rect(x, 25, 10, 10, 1)
    oled.show()
    
    time.sleep(0.01) # Small delay to control speed
```
