So far, we have only sent signals **out** of the Pico. Now we need to **listen** to signals coming **in**.

A Digital Input is the Pico's way of knowing if a circuit has been closed or opened (High vs. Low).

## 1. The Floating State Problem

When a pin is not connected to anything, it is "floating." A floating pin picks up static electricity and noise from the air, causing it to flip between 0 and 1 randomly. 

To fix this, we use **Pull-up** or **Pull-down** resistors to "tie" the pin to a known state (0V or 3.3V) when no signal is present.

```python
import machine
import time

# Setup a pin as an INPUT with an internal PULL_DOWN resistor
# This means the pin will be 0 (LOW) by default.
input_pin = machine.Pin(14, machine.Pin.IN, machine.Pin.PULL_DOWN)

print("System listening on GP14...")
```

## 2. Reading Binary States

We use `.value()` to check the current state of a pin. 
- **1**: The pin is receiving 3.3V (Logic HIGH).
- **0**: The pin is connected to Ground (Logic LOW).

```python
while True:
    state = input_pin.value()
    if state == 1:
        print("Event Detected! State: HIGH", end="\r")
    else:
        print("System Idle... State: LOW  ", end="\r")
    
    time.sleep(0.1)
```

## 3. Momentary vs. Toggle Logic

In software, we can decide how to interpret a signal. 

1. **Momentary**: Something happens only while the signal is HIGH.
2. **Toggle**: A signal flip-flops the state of the system permanently (e.g., ON -> OFF).

```python
system_active = False

while True:
    if input_pin.value() == 1:
        # Toggle the system state
        system_active = not system_active
        print(f"System toggled. Active: {system_active}")
        
        # Wait for the signal to end before proceeding
        # This prevents the system from toggling 100 times in one second
        while input_pin.value() == 1:
            time.sleep(0.01)
            
    time.sleep(0.05)
```
