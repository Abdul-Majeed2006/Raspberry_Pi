## Digital vs. Analog

![Analog vs Digital](assets/adc_analog_signals_visual.png)

- **Digital**: Only two values. 0V or 3.3V. (No in-between).
- **Analog**: Infinite values between 0V and 3.3V. (e.g., 1.5V, 2.78V). 

The Pico has special translators called **ADCs (Analog-to-Digital Converters)**.

## 1. The Internal Temperature Sensor

Did you know your Pico has a thermometer inside it? It's connected to **ADC Channel 4**. 
Let's read it.

The ADC gives us a number between 0 and 65535. We have to do some math to convert that into Degrees Celsius.

```python
import machine
import time

# Setup the ADC on the special internal channel (4)
sensor_temp = machine.ADC(4)
conversion_factor = 3.3 / 65535

while True:
    # 1. Read the raw value (0-65535)
    reading = sensor_temp.read_u16()
    
    # 2. Convert to Voltage (0V - 3.3V)
    voltage = reading * conversion_factor
    
    # 3. Convert Voltage to Temperature (Formula from the Datasheet)
    temperature = 27 - (voltage - 0.706) / 0.001721
    
    print(f"Temp: {temperature:.1f} C")
    time.sleep(1)
```

## 2. Reading a Potentiometer (Knob)

If you have a potentiometer (a twisting knob), this is where the fun begins. 
**Wiring:**
- Left Leg -> 3.3V
- Right Leg -> GND
- Middle Leg (Wiper) -> **GP26 (ADC0)**

*Check your Pinout Image again to find GP26!*

```python
# Reading the Knob
potentiometer = machine.ADC(26) # GP26 is ADC0

while True:
    val = potentiometer.read_u16()
    print(f"Knob Value: {val}")
    time.sleep(0.1)
```

## 3. Project: The Dimmer Switch

Let's map the Knob value directly to the LED brightness.
Since both use 0-65535, it's a perfect match!

```python
# The Perfect Dimmer
led = machine.PWM(machine.Pin("LED", machine.Pin.OUT))
led.freq(1000)

pot = machine.ADC(26)

while True:
    val = pot.read_u16()
    led.duty_u16(val)
    # No sleep needed really, instant response!
```
