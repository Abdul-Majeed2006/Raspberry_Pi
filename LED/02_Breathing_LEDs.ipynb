In Chapter 1, our world was binary. Light ON. Light OFF. 
But the real world has *gradients*. Dim lights, smooth transitions, sunsets.

Computers (and the Pico) are strictly digital. They only know 0V and 3.3V. So how do we dim an LED? We cheat.

## Enter PWM (Pulse Width Modulation)

Imagine switching a light switch on and off really, really fast. 
- If it's ON 50% of the time and OFF 50% of the time, your brain averages it out and sees "50% brightness".
- If it's ON 10% of the time, it looks dim.
- If it's ON 90% of the time, it looks bright.

![PWM Concept](assets/pwm_concept_diagram.png)

This trick is called **Pulse Width Modulation (PWM)**.

## 1. Setting Up PWM

We need the `machine` library again, but this time we wrap our Pin in a `PWM` object.

```python
import machine
import time

# Setup the onboard LED (or an external one on GP15 if you have it)
led = machine.Pin("LED", machine.Pin.OUT)
pwm_led = machine.PWM(led)

# Frequency: How fast we switch on/off. 
# 1000 Hz = 1000 times per second. Fast enough that our eyes can't see the flicker.
pwm_led.freq(1000)
```

## 2. Controlling Brightness (Duty Cycle)

In MicroPython, the brightness (Duty Cycle) is a 16-bit number. 
- **0** = 0% (OFF)
- **65535** = 100% (Full ON)
- **32768** â‰ˆ 50%

Let's try setting it to different levels.

```python
# Set to ~10% brightness
pwm_led.duty_u16(6500)
print("Dim glow...")
time.sleep(2)

# Set to 50% brightness
pwm_led.duty_u16(32768)
print("Half power!")
time.sleep(2)

# Full blast
pwm_led.duty_u16(65535)
print("MAXIMUM POWER!")
time.sleep(2)

# Turn off
pwm_led.duty_u16(0)
```

## 3. The "Breathing" Effect

To make the LED look like it's breathing, we need to smoothly increase the brightness from 0 to 65535, and then decrease it back to 0.

```python
import time

print("Breathing started... (Press Stop to end)")

while True:
    # Brighter
    for duty in range(0, 65535, 1000):
        pwm_led.duty_u16(duty)
        time.sleep(0.01)
    
    # Dimmer
    for duty in range(65535, 0, -1000):
        pwm_led.duty_u16(duty)
        time.sleep(0.01)
```
