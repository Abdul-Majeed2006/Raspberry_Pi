## The Game Logic

![Reaction Logic](assets/game_reaction_timer_logic.png)

1. **Attract Mode**: Flash lights to invite player.
2. **Wait**: Random delay (to prevent cheating).
3. **GO!**: Turn on the light.
4. **Measure**: Count milliseconds until button press.
5. **Result**: 
    - < 200ms: SUPERHUMAN (Green)
    - 200-500ms: AVERAGE (Yellow/Blue)
    - > 500ms: SLOW (Red)

## 1. Setup Your Rig

- **Button**: GP14
- **RGB LED**: GP16 (R), GP17 (G), GP18 (B)

```python
import machine
import time
import random

# --- SETUP ---
button = machine.Pin(14, machine.Pin.IN, machine.Pin.PULL_DOWN)

led_r = machine.Pin(16, machine.Pin.OUT)
led_g = machine.Pin(17, machine.Pin.OUT)
led_b = machine.Pin(18, machine.Pin.OUT)

def set_color(r, g, b):
    led_r.value(r)
    led_g.value(g)
    led_b.value(b)

print("Game System Initialized.")
```

## 2. The Game Loop

We will run this inside a `while True` loop so you can play over and over.

```python
print("Press Ctrl+C (Stop) to exit.")

while True:
    # --- 1. ATTRACT MODE ---
    print("\n--- NEW GAME ---")
    set_color(1, 0, 1) # Purple waiting state
    
    # Wait for player to be ready (press button to start)
    print("Press button to start...")
    while button.value() == 0:
        time.sleep(0.01)
    
    # Wait for release
    while button.value() == 1:
        time.sleep(0.01)
        
    # --- 2. RANDOM WAIT ---
    set_color(0, 0, 0) # Black out
    print("Get Ready...")
    
    # Wait between 2 and 5 seconds randomly
    delay = random.uniform(2, 5)
    time.sleep(delay)
    
    # Check for early start (Cheating!)
    if button.value() == 1:
        print("TOO EARLY! Disqualified.")
        for i in range(5):
            set_color(1, 0, 0) # Red Flash
            time.sleep(0.1)
            set_color(0, 0, 0)
            time.sleep(0.1)
        continue # Skip to restart

    # --- 3. GO! ---
    set_color(1, 1, 1) # WHITE FLASH
    start_time = time.ticks_ms() # Record exact time in milliseconds
    
    # Wait for reaction
    while button.value() == 0:
        pass # Do nothing, just wait fast
        
    # As soon as loop breaks (button pressed):
    reaction_time = time.ticks_diff(time.ticks_ms(), start_time)
    print(f"Reaction Time: {reaction_time} ms")
    
    # --- 4. SCORING ---
    if reaction_time < 200:
        print("RANK: LEGENDARY âš¡")
        set_color(0, 1, 0) # Green
    elif reaction_time < 350:
        print("RANK: HUMAN ðŸ˜")
        set_color(0, 0, 1) # Blue
    else:
        print("RANK: SLOTH ðŸ¢")
        set_color(1, 0, 0) # Red
        
    # Show score for 3 seconds
    time.sleep(3)
```
